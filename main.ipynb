2.Decomposition of trend and seasonality

Seasonal decomposition breaks a time series into three components:
1.Trend  the long-term overall direction (upward or downward).
2.Seasonality  repeating patterns at regular intervals (e.g., weekly, monthly).
3.Residual (or noise)  random fluctuations not explained by trend or seasonality.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['Local Route'], model='additive', period=30)
result.plot()
plt.show()
```

3.Rolling Window 

A rolling mean (or rolling average) smooths a time series by averaging data points over a fixed window (e.g., 7 days). It reduces short-term fluctuations and highlights long-term trends or patterns in the data.

```python
rolling_window = 7
df_rolling = df.rolling(window=rolling_window).mean()

plt.figure(figsize=(14,6))
for col in df.columns:
    plt.plot(df_rolling[col], label=col)
plt.title("7-Day Rolling Average of Passenger Counts")
plt.xlabel("Date")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

FORECASTING

```python
from prophet import Prophet
import pandas as pd

# Create a copy and reset index
df_copy = df.copy().reset_index()  # 'Date' becomes a column again

# List of services to forecast
services = ['Local Route','Light Rail','Peak Service','Rapid Route','School']
forecasts = {}

# Loop over each service
for service in services:
    # Prepare data for Prophet
    df_service = df_copy[['Date', service]].rename(columns={'Date':'ds', service:'y'})
    df_service['y'] = df_service['y'].clip(lower=0)
    
    # Initialize and fit Prophet model
    model = Prophet(weekly_seasonality=True)
    model.fit(df_service)
    
    # Create future dataframe for next 7 days
    future = model.make_future_dataframe(periods=7)
    forecast = model.predict(future)
    
    # Clip negative predictions to 0
    forecast['yhat'] = forecast['yhat'].clip(lower=0)
    

    # Store ONLY next 7 days (tail)
    forecasts[service] = forecast[['ds','yhat']].tail(7)

```

```python
df.tail(7)
```

```python
for service in services:
    print(f"Forecast for {service}:")
    print(forecasts[service])
    print("\n")
```

1.All services show strong weekday demand and steep weekend drops.
2.School and Peak services go to zero during weekends  clear pattern.
3.Local Route, Rapid Route, and Light Rail still operate on weekends but at very low demand.

Weekly seasonality

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

services = ['Local Route','Light Rail','Peak Service','Rapid Route','School']
forecasts = {}
metrics = {}

for service in services:
    print(f"\n----- Evaluating: {service} -----")

    # Prepare data
    df_service = df_copy[['Date', service]].rename(columns={'Date':'ds', service:'y'})
    df_service['y'] = df_service['y'].clip(lower=0)

    # ------------------------------
    # 1. TRAIN-TEST SPLIT (last 30 days as test)
   
    test_size = 30
    train = df_service[:-test_size]
    test  = df_service[-test_size:]

    
    # 2. Train Prophet on train data
  
    model = Prophet(weekly_seasonality=True)
    model.fit(train)

  
    # Predict on test period
  
    future = model.make_future_dataframe(periods=test_size)
    forecast = model.predict(future)

    pred = forecast[['ds','yhat']].tail(test_size)

    pred['yhat'] = pred['yhat'].clip(lower=0)

 
    y_true = test['y'].values
    y_pred = pred['yhat'].values

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))

    print("MAE:", mae)
    print("RMSE:", rmse)


    metrics[service] = {'MAE': mae, 'RMSE': rmse}

  
    future_7 = model.make_future_dataframe(periods=7)
    forecast_7 = model.predict(future_7)
    forecast_7['yhat'] = forecast_7['yhat'].clip(lower=0)

    forecasts[service] = forecast_7[['ds','yhat']].tail(7)
```
